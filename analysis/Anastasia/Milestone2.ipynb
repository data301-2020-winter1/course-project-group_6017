#### Plot 1: Count Plot
- Shows a count of obervations for a category. I will use count plots to show the if the rocket launches have been state run or privately funded from 1964-2020 and to show which countries have launched the most rockets

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scripts import project_functions
df = project_functions.load_and_process("Global Space Launches.csv")
sns.set_theme(style="white",font_scale=1.3)
fig, ax=plt.subplots(figsize=(30,8))
ax = sns.countplot(x='Year', hue="Private or State Run",  data=df)
plt.title('Rocket launches Private or State run over time')
sns.despine()
plt.show()
```

#### Plot 2: Histplot
- I used a histplot to show the distibution between private or state run rocket launches seperated by country 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scripts import project_functions
df = project_functions.load_and_process("Global Space Launches.csv")
sns.set_theme(style="white",font_scale=1.3)
fig, ax=plt.subplots(figsize=(15,10))
ax = sns.histplot(x="Companys Country of Origin", hue="Private or State Run", data=df, multiple="dodge", palette='rocket')
sns.despine()
plt.title('Private or State Run by Country')
ax.set(xlabel='P or S in Country', ylabel='Number of Rockets')
plt.show()
```

#### Plot 3: Scatterplot
- I used a scatterplot to show which company has launched a rocket in the span of 1964-2020 and wether or not they are still active. I made another to also show the success rate of each launch.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scripts import project_functions
df = project_functions.load_and_process("Global Space Launches.csv")
sns.set_theme(style="white",font_scale=1.3)
fig, ax=plt.subplots(figsize=(15,10))
ax = sns.scatterplot(x='Year', y='Company Name', hue='Status Rocket', data=df)
sns.despine()
plt.title('Rocket Launches per Company over time')
ax.set(xlabel='Year', ylabel='Company Name')
plt.show()
```
